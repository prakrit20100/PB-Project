```python
from scipy.spatial.distance import cdist
distortions = []
inertias = []
mapping1 = {}
mapping2 = {}

for i in range(1,10):
    
    kmeanModel = KMeans(n_clusters=i).fit(tsne1)
    kmeanModel.fit(tsne1)
 
    distortions.append(sum(np.min(cdist(tsne1, kmeanModel.cluster_centers_,
                                        'euclidean'), axis=1)) / tsne1.shape[0])
    inertias.append(kmeanModel.inertia_)
 
    mapping1[i] = sum(np.min(cdist(tsne1, kmeanModel.cluster_centers_,
                                   'euclidean'), axis=1)) / tsne1.shape[0]
    mapping2[i] = kmeanModel.inertia_
```

```python
for key, val in mapping1.items():
    print(f'{key} : {val}')
```

```python
for key, val in mapping2.items():
    print(f'{key} : {val}')
```

```python
K = range(1,10)
plt.plot(K, inertias, 'bx-')
plt.xlabel('Values of K')
plt.ylabel('Inertia')
plt.title('The Elbow Method using Inertia')
plt.show()
```
